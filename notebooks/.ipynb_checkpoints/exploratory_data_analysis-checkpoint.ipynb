```python
# ------------------------------------------------------
# 1. Load data
# ------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)


tournament_palette = sns.color_palette("viridis", n_colors=6)

squads = pd.read_csv('../data-processed/squads_clean.csv')
substitutions = pd.read_csv('../data-processed/substitutions_clean.csv')
bookings = pd.read_csv('../data-processed/bookings_clean.csv')
goals = pd.read_csv('../data-processed/goals_clean.csv')
matches_weather = pd.read_csv('../data-processed/matches_clean_weather.csv')
matchbox = pd.read_csv('../data-processed/MatchBoxData_clean.csv')

print("All files loaded successfully!")

# Dictionary to store and display column names for each df
datasets = {
    "Squads": squads,
    "Substitutions": substitutions,
    "Bookings": bookings,
    "Goals": goals,
    "Matches Weather": matches_weather,
    "MatchBox Data": matchbox
}

for name, df in datasets.items():
    print(f"--- {name} Overview ---")
    df.info()
    print("\n" + "="*50 + "\n")
```

```python
# ------------------------------------------------------
# 2. Feature engineering
# ------------------------------------------------------
# 1. Avg weather variables per match
# Mean of the 3 hourly readings (start, +1hr, +2hr)
matches_weather['avg_temp'] = matches_weather[['temp0', 'temp1', 'temp2']].mean(axis=1)
matches_weather['avg_humid'] = matches_weather[['humid0', 'humid1', 'humid2']].mean(axis=1)
matches_weather['avg_wind'] = matches_weather[['wind0', 'wind1', 'wind2']].mean(axis=1)

# 2. Total goals per match
# Group by match_id to count how many goals happened in each game
goals_per_match = goals.groupby('match_id').size().reset_index(name='total_goals')

# 3. Merge total goals into main match df
# Use left merge to keep all matches, filling 0 for 0-0 draws if any are missing from goals df
matches_full = pd.merge(matches_weather, goals_per_match, on='match_id', how='left')
matches_full['total_goals'] = matches_full['total_goals'].fillna(0)

# 4. Extract year from tournament ID for cleaner sorting
print("Data preprocessing complete.")
matches_full.head(3)
```

```python
# ------------------------------------------------------
# 3. Goal distribution
# ------------------------------------------------------
# Histogram of goals per match
plt.figure(figsize=(10, 6))
sns.countplot(
    data=matches_full, 
    x='total_goals', 
    hue='total_goals',     
    palette='Blues_d', 
    legend=False           
)

plt.title('Distribution of Goals per Match (All Tournaments)', fontsize=16)
plt.xlabel('Number of Goals', fontsize=12)
plt.ylabel('Count of Matches', fontsize=12)

# Calculate mean goals
mean_goals = matches_full['total_goals'].mean()
plt.axvline(mean_goals, color='red', linestyle='--', label=f'Mean Goals: {mean_goals:.2f}')
plt.legend()
plt.show()

print(f"Skewness of goals: {matches_full['total_goals'].skew()}")
```

```python
# Right-skewed distribution (skewness ~0.78) with an average of 2.51 goals 
# per match, most games feature 2-3 goals while high-scoring matches are 
# rare outliers. It resembles a Poisson distribution.
```

```python
# ------------------------------------------------------
# 4. Goals vs. Tournament
# ------------------------------------------------------
plt.figure(figsize=(12, 6))
# Boxplot to see the spread (median, quartiles) per tournament
sns.boxplot(data=matches_full, x='tournament_id', hue='tournament_id', y='total_goals', palette="Set2", legend=False)

plt.title('Goals per Match Distribution by Tournament', fontsize=16)
plt.xlabel('Tournament', fontsize=12)
plt.ylabel('Goals Scored', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Goals seems largely consistent across the tournaments, notable exception would
# be 2014, with higher median of 3, and greater variability
```

```python
# ------------------------------------------------------
# 5. Timing of goals, substitutions, and yellow cards
# ------------------------------------------------------
plt.figure(figsize=(14, 7))

# Plot 1: Goal timings
sns.kdeplot(goals['minute_regulation'], label='Goals', fill=True, color='green', alpha=0.3)

# Plot 2: Substitution timings
sns.kdeplot(substitutions['minute_regulation'], label='Substitutions', fill=True, color='#673AB7', alpha=0.3)

# Plot 3: Yellow card timings
# Filter only yellow cards
yellow_cards = bookings[bookings['is_yellow'] == True]
sns.kdeplot(yellow_cards['card_minute'], label='Yellow Cards', fill=True, color='yellow', alpha=0.3)

plt.title('Timeline Density: Goals vs. Subs vs. Cards', fontsize=16)
plt.xlabel('Match Minute', fontsize=12)
plt.ylabel('Density (Frequency)', fontsize=12)
plt.xlim(0, 120) # Limit to typical match time + extra time
plt.legend()
plt.show()
```

```python
# ------------------------------------------------------
# 6. Temperature distribution
# ------------------------------------------------------
plt.figure(figsize=(14, 6))

sns.violinplot(data=matches_full, x='tournament_id', hue='tournament_id', y='avg_temp', inner="quart", palette="coolwarm", legend=False)

plt.title('Temperature Distributions by Tournament', fontsize=16)
plt.ylabel('Average Match Temperature (Â°C)', fontsize=12)
plt.xlabel('Tournament', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------------------------------------------
# 7. Squad median age across tournaments
# ------------------------------------------------------
# 1. Convert rank_date to datetime objects
squads['rank_date'] = pd.to_datetime(squads['rank_date'], format='%Y-%m-%d')

# 2. Extract year
squads['tournament_year'] = squads['rank_date'].dt.year

# 3. Plot squad ages by tournament year
plt.figure(figsize=(12, 6))

sns.boxplot(
    data=squads, 
    x='tournament_year',
    hue='tournament_year',
    y='median_age', 
    palette="Purples",
    legend=False
)

plt.title('Distribution of Squad Median Ages by Tournament', fontsize=16)
plt.xlabel('Tournament Year', fontsize=12)
plt.ylabel('Median Squad Age', fontsize=12)
plt.show()
```

```python
# ------------------------------------------------------
# 8. Comparing red cards in group stage vs. knockout
# ------------------------------------------------------
logical_order = [
    "group stage", 
    "round of 16", 
    "quarter-finals", 
    "semi-finals", 
    "third-place match", 
    "final"
]
# 1. Total matches per stage per tournament
stage_match_counts = matches_weather.groupby(['tournament_id', 'stage_name'])['match_id'].nunique().reset_index(name='total_matches')

# 2. Red cards per stage per tournament
stage_red_cards = bookings[bookings['is_red'] == True].groupby(['tournament_id', 'stage_name']).size().reset_index(name='total_red_cards')

# 3. Merge dfs
# Left merge on match counts to ensure we keep stages even if they had 0 red cards
stage_stats = pd.merge(stage_match_counts, stage_red_cards, on=['tournament_id', 'stage_name'], how='left')

# Fill NaN values with 0 (for stages that had 0 red cards)
stage_stats['total_red_cards'] = stage_stats['total_red_cards'].fillna(0)

# 4. Calculate the metric
stage_stats['red_cards_per_match'] = stage_stats['total_red_cards'] / stage_stats['total_matches']

# 5. Plotting
plt.figure(figsize=(12, 6))
sns.barplot(
    data=stage_stats, 
    x='stage_name', 
    y='red_cards_per_match', 
    hue='tournament_id', 
    palette='Reds',
    order=logical_order
)

plt.title('Normalised Red Card Rate (Cards per Match)', fontsize=16)
plt.xlabel('Stage', fontsize=12)
plt.ylabel('Red Cards per Match', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Tournament', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# ------------------------------------------------------
# 9. Correlation heatmap of match stats
# ------------------------------------------------------
# 1. Prepare substitution counts
subs_per_match = substitutions.groupby('match_id').size().reset_index(name='total_subs')

# 2. Create "Master Stats" df
# Start with MatchBox data
match_stats = matchbox.copy()

# 3. Merge with Weather & Goals using Team Names + Tournament
# Use this merge to fetch the match_id, avg_temp, and total_goals
match_stats = pd.merge(
    match_stats, 
    matches_full[['tournament_id', 'home_team_name', 'away_team_name', 'match_id', 'avg_temp', 'total_goals']], 
    left_on=['tournament_id', 'hname', 'aname'],       # Keys in MatchBox
    right_on=['tournament_id', 'home_team_name', 'away_team_name'], # Keys in Matches Full
    how='left'
)

# 4. Merge the Subs
match_stats = pd.merge(match_stats, subs_per_match, on='match_id', how='left')
match_stats['total_subs'] = match_stats['total_subs'].fillna(0)

# 5. Calculate match totals
match_stats['total_shots'] = match_stats['hshots'] + match_stats['ashots']
match_stats['total_fouls'] = match_stats['hfouls'] + match_stats['afouls']
match_stats['total_yellows'] = match_stats['hyellowCards'] + match_stats['ayellowCards']
match_stats['total_reds'] = match_stats['hredCards'] + match_stats['aredCards']

# 6. Select columns
cols_of_interest = [
    'total_goals', 
    'total_shots', 
    'total_fouls', 
    'total_yellows', 
    'total_reds', 
    'total_subs', 
    'avg_temp',
    'hPossession'
]

readable_names = {
    'total_goals': 'Total Goals',
    'total_shots': 'Total Shots',
    'total_fouls': 'Fouls',
    'total_yellows': 'Yellow Cards',
    'total_reds': 'Red Cards',
    'total_subs': 'Substitutions',
    'avg_temp': 'Temperature',
    'hPossession': 'Home Possession'
}

# Calculate correlation on the renamed df
corr_matrix = match_stats[cols_of_interest].rename(columns=readable_names).corr()

# 7. Plotting
plt.figure(figsize=(10, 8))

# Mask upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix, 
    mask=mask, 
    annot=True, 
    fmt=".2f", 
    cmap='vlag', 
    center=0,
    vmin=-0.5, 
    vmax=0.5,
    square=True, 
    linewidths=1, 
    linecolor='white',
    cbar_kws={"shrink": .7},
    annot_kws={"size": 11} 
)

plt.title('Correlation of Match Intensity & Conditions', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right') # Rotate x-labels for readability
plt.yticks(rotation=0)
plt.show()
```

```python
# ------------------------------------------------------
# 11. Impact of FIFA Ranking on match outcome 
# ------------------------------------------------------
# 1. Prepare rank data 
matches_full['tournament_year'] = matches_full['tournament_id'].apply(lambda x: int(x.split('-')[1]))

# Merge ranks
matches_ranked = pd.merge(
    matches_full, 
    squads[['country_full', 'tournament_year', 'rank']], 
    left_on=['home_team_name', 'tournament_year'], 
    right_on=['country_full', 'tournament_year'], 
    how='left'
).rename(columns={'rank': 'home_rank'})

matches_ranked = pd.merge(
    matches_ranked, 
    squads[['country_full', 'tournament_year', 'rank']], 
    left_on=['away_team_name', 'tournament_year'], 
    right_on=['country_full', 'tournament_year'], 
    how='left'
).rename(columns={'rank': 'away_rank'})

# 2. Calculate differentials
matches_ranked['rank_diff'] = matches_ranked['away_rank'] - matches_ranked['home_rank']
matches_ranked['score_margin'] = matches_ranked['home_team_score'] - matches_ranked['away_team_score']

# 3. Define result category for colouring
def get_result_type(row):
    if row['score_margin'] > 0: return 'Home Win'
    elif row['score_margin'] < 0: return 'Away Win'
    else: return 'Draw'

matches_ranked['result_type'] = matches_ranked.apply(get_result_type, axis=1)

# 4. Plotting
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=matches_ranked, 
    x='rank_diff', 
    y='score_margin', 
    hue='result_type', 
    palette={'Home Win': 'mediumseagreen', 'Away Win': 'indianred', 'Draw': 'lightgrey'},
    alpha=0.6,
    s=60
)

# Add regression line 
sns.regplot(
    data=matches_ranked, 
    x='rank_diff', 
    y='score_margin', 
    scatter=False, 
    line_kws={'color':'black', 'linestyle':'--', 'linewidth':1.5}
)

# 5. Annotations & layout
plt.title('FIFA Rankings vs. Match Outcome', fontsize=16)
plt.xlabel('Home Rank Advantage', fontsize=12)
plt.ylabel('Goal Margin', fontsize=12)

# Add quadrant guidelines
plt.axhline(0, color='black', linewidth=0.5, alpha=0.5)
plt.axvline(0, color='black', linewidth=0.5, alpha=0.5)

# Add quadrant labels
# Top-Right
plt.text(40, 4, "Expected\nHome Win", fontsize=11, color='green', fontweight='bold', ha='center')
# Bottom-Left
plt.text(-40, -4, "Expected\nAway Win", fontsize=11, color='darkred', fontweight='bold', ha='center')
# Top-Left
plt.text(-40, 4, "HOME\nUPSET", fontsize=10, color='grey', fontweight='bold', ha='center')
# Bottom-Right
plt.text(40, -4, "AWAY\nUPSET", fontsize=10, color='grey', fontweight='bold', ha='center')
plt.legend(title='Match Result')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# ------------------------------------------------------
# 12. Most efficient teams by shot conversion
# ------------------------------------------------------
# 1. Aggregate stats by home team
home_stats = matchbox.groupby('hname')[['hshots', 'hshotsOnTarget']].sum()
home_goals = matches_full.groupby('home_team_name')['home_team_score'].sum()

# 2. Combine and calculate
efficiency = pd.concat([home_stats, home_goals], axis=1)
efficiency.columns = ['Shots', 'ShotsOnTarget', 'Goals']
efficiency = efficiency.fillna(0)

# Filter for teams that played at least a few games (e.g., > 20 shots total) to avoid outliers
efficiency = efficiency[efficiency['Shots'] > 20]

# 3. Conversion rate
efficiency['conversion_rate'] = (efficiency['Goals'] / efficiency['Shots']) * 100

# 4. Plot top 10 clinical teams
top_10 = efficiency.sort_values('conversion_rate', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=top_10['conversion_rate'], 
    y=top_10.index, 
    hue=top_10.index,
    palette='magma',
    legend=False
)

plt.title('Top 10 Most Clinical Teams (Highest Goal/Shot Ratio)', fontsize=16)
plt.xlabel('Conversion Rate (%)', fontsize=12)
plt.ylabel('Team', fontsize=12)

# Annotate values
for index, value in enumerate(top_10['conversion_rate']):
    plt.text(value + 0.5, index, f"{value:.1f}%", va='center')

plt.show()
```

```python
# ------------------------------------------------------
# 13.1 Leaderboards
# ------------------------------------------------------
# 1. Convert match rows to team rows
def create_team_stats(matchbox_df, weather_df):
    stats_cols = ['tournament_id', 'hname', 'aname', 'hPossession', 'aPossession', 
                  'hshots', 'ashots', 'hshotsOnTarget', 'ashotsOnTarget', 
                  'hfouls', 'afouls', 'hyellowCards', 'ayellowCards', 'hredCards', 'aredCards']
    
    # Bring in score, extra time, and penalties from weather/matches file
    base_df = pd.merge(
        matchbox_df[stats_cols],
        weather_df[['tournament_id', 'home_team_name', 'away_team_name', 
                   'home_team_score', 'away_team_score', 'extra_time', 'penalty_shootout']],
        left_on=['tournament_id', 'hname', 'aname'],
        right_on=['tournament_id', 'home_team_name', 'away_team_name'],
        how='inner'
    )

    # --- Container for long list ---
    team_rows = []

    for _, row in base_df.iterrows():
        home_data = {
            'tournament_id': row['tournament_id'],
            'team_name': row['hname'],
            'goals': row['home_team_score'],
            'goals_conceded': row['away_team_score'],
            'shots': row['hshots'],
            'shots_on_target': row['hshotsOnTarget'],
            'possession': row['hPossession'],
            'fouls': row['hfouls'],
            'yellow_cards': row['hyellowCards'],
            'red_cards': row['hredCards'],
            'is_extra_time': row['extra_time'],
            'is_penalty': row['penalty_shootout'],
            # Win Calculation: Score > Opponent
            'is_win': 1 if row['home_team_score'] > row['away_team_score'] else 0
        }
        
        away_data = {
            'tournament_id': row['tournament_id'],
            'team_name': row['aname'],
            'goals': row['away_team_score'],
            'goals_conceded': row['home_team_score'],
            'shots': row['ashots'],
            'shots_on_target': row['ashotsOnTarget'],
            'possession': row['aPossession'],
            'fouls': row['afouls'],
            'yellow_cards': row['ayellowCards'],
            'red_cards': row['aredCards'],
            'is_extra_time': row['extra_time'],
            'is_penalty': row['penalty_shootout'],
            'is_win': 1 if row['away_team_score'] > row['home_team_score'] else 0
        }
        
        team_rows.append(home_data)
        team_rows.append(away_data)

    return pd.DataFrame(team_rows)

# 2. Run transformation
# Only need matchbox and matches_weather now
squad_match_stats = create_team_stats(matchbox, matches_weather)

# 3. Aggregate by squad (team + tournament)
squad_leaderboard = squad_match_stats.groupby(['team_name', 'tournament_id']).agg(
    matches_played=('is_win', 'count'),
    total_wins=('is_win', 'sum'),
    avg_goals=('goals', 'mean'),
    avg_shots=('shots', 'mean'),
    avg_shots_on_target=('shots_on_target', 'mean'),
    avg_possession=('possession', 'mean'),
    avg_fouls=('fouls', 'mean'),
    avg_yellows=('yellow_cards', 'mean'),
    avg_reds=('red_cards', 'mean'),
    avg_extra_time=('is_extra_time', 'mean'), # % of matches going to extra-time
    avg_penalties=('is_penalty', 'mean')      # % of matches going to penalty shoot-out
).reset_index()

# 4. Create squad label
squad_leaderboard['year'] = squad_leaderboard['tournament_id'].apply(lambda x: x.split('-')[1] if '-' in str(x) else x)
squad_leaderboard['squad_label'] = squad_leaderboard['team_name'] + " " + squad_leaderboard['year']

# 5. Merge with median age and rank
squad_leaderboard['year_int'] = squad_leaderboard['year'].astype(int)

final_leaderboard = pd.merge(
    squad_leaderboard,
    squads[['country_full', 'tournament_year', 'median_age', 'rank']],
    left_on=['team_name', 'year_int'],
    right_on=['country_full', 'tournament_year'],
    how='left'
)

# 6. Calculate win %
final_leaderboard['win_pct'] = (final_leaderboard['total_wins'] / final_leaderboard['matches_played']) * 100

final_leaderboard.head()
```

```python
# ------------------------------------------------------
# 13.2 Leaderboards
# ------------------------------------------------------
import math

# Metrics to plot (dictionary: {'Column Name': 'Display Title'})
metrics = {
    'win_pct': 'Highest Win Rate (%)',
    'avg_goals': 'Goals per Match (Avg)',
    'median_age': 'Oldest Squads (Median Age)', # For youngest, we'll reverse sort
    'avg_shots': 'Shots per Match (Avg)',
    'avg_shots_on_target': 'Shots on Target (Avg)',
    'avg_possession': 'Possession % (Avg)',
    'avg_fouls': 'Most Aggressive (Fouls/Match)',
    'avg_yellows': 'Yellow Cards per Match',
    'avg_reds': 'Red Cards per Match',
    'avg_extra_time': 'Most Extra-Time Games (%)',
    'avg_penalties': 'Most Penalty Shootouts (%)'
}

# Setup plot grid
n_metrics = len(metrics) + 2 # +1 for Youngest, +1 for Avg Rank (Total = 13)

cols = 3
rows = 5 
fig, axes = plt.subplots(rows, cols, figsize=(18, 25)) 
axes = axes.flatten()

colors = ['#FFD700', '#C0C0C0', '#CD7F32', '#4682B4', '#5F9EA0'] # Gold, Silver, Bronze, Blue, Teal

def plot_top5(ax, data, col, title, ascending=False):
    # Sort and take top 5
    top5 = data.sort_values(col, ascending=ascending).head(5)
    
    # Plot horizontal bar
    sns.barplot(
        data=top5,
        x=col,
        y='squad_label',
        hue='squad_label', 
        ax=ax,
        palette=colors,
        edgecolor='black',
        legend=False      
    )
    
    # Styling
    ax.set_title(title, fontsize=14, fontweight='bold', pad=10)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=11, labelcolor='#333333')
    
    # Add values to end of bars
    for i, v in enumerate(top5[col]):
        val_str = f"{v:.1f}" if isinstance(v, float) else str(v)
        ax.text(v, i, f" {val_str}", va='center', fontweight='bold')

# --- Loop through standard metrics ---
for i, (col, title) in enumerate(metrics.items()):
    # Filter: Only include teams that played at least 3 matches to avoid "1 game wonders"
    valid_squads = final_leaderboard[final_leaderboard['matches_played'] >= 3]
    
    # Handle specific sort orders
    is_ascending = False # Default: Higher is better
    
    plot_top5(axes[i], valid_squads, col, title, ascending=is_ascending)

# --- Special chart 1: youngest squads (reverse of oldest) ---
valid_squads = final_leaderboard[final_leaderboard['matches_played'] >= 3]
plot_top5(axes[len(metrics)], valid_squads, 'median_age', 'Youngest Squads (Median Age)', ascending=True)

# --- Special chart 2: highest avg ranking (minimum participation filter) ---
# 1. Count how many tournaments each team played in
participation_counts = final_leaderboard['team_name'].value_counts()

# 2. Only keep teams that played in at least 3 tournaments
valid_teams = participation_counts[participation_counts >= 3].index
filtered_leaderboard = final_leaderboard[final_leaderboard['team_name'].isin(valid_teams)]

# 3. Aggregate
country_rank = filtered_leaderboard.groupby('team_name')['rank'].mean().reset_index()
country_rank['squad_label'] = country_rank['team_name'] 

# 4. Plot
plot_top5(axes[len(metrics)+1], country_rank, 'rank', 'Best Avg FIFA Ranking (All Eras)', ascending=True)

# Clean up empty subplots (indices 13 and 14)
for i in range(n_metrics, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.suptitle('Top 5 Leaderboards (2002-2022)', fontsize=24, y=1.01, fontweight='heavy')
plt.show()
```
