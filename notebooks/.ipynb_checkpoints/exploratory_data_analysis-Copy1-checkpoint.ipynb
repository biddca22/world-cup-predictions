```python
# ------------------------------------------------------
# 1. Load data
# ------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)


tournament_palette = sns.color_palette("viridis", n_colors=6)

squads = pd.read_csv('../data-processed/squads_clean.csv')
substitutions = pd.read_csv('../data-processed/substitutions_clean.csv')
bookings = pd.read_csv('../data-processed/bookings_clean.csv')
goals = pd.read_csv('../data-processed/goals_clean.csv')
matches_weather = pd.read_csv('../data-processed/matches_clean_weather.csv')
matchbox = pd.read_csv('../data-processed/MatchBoxData_clean.csv')

print("All files loaded successfully!")

# Dictionary to store and display column names for each df
datasets = {
    "Squads": squads,
    "Substitutions": substitutions,
    "Bookings": bookings,
    "Goals": goals,
    "Matches Weather": matches_weather,
    "MatchBox Data": matchbox
}

for name, df in datasets.items():
    print(f"--- {name} Overview ---")
    df.info()
    print("\n" + "="*50 + "\n")
```

```python
# ------------------------------------------------------
# 2. Feature engineering
# ------------------------------------------------------
# 1. Avg weather variables per match
# Mean of the 3 hourly readings (start, +1hr, +2hr)
matches_weather['avg_temp'] = matches_weather[['temp0', 'temp1', 'temp2']].mean(axis=1)
matches_weather['avg_humid'] = matches_weather[['humid0', 'humid1', 'humid2']].mean(axis=1)
matches_weather['avg_wind'] = matches_weather[['wind0', 'wind1', 'wind2']].mean(axis=1)

# 2. Total goals per match
# Group by match_id to count how many goals happened in each game
goals_per_match = goals.groupby('match_id').size().reset_index(name='total_goals')

# 3. Merge total goals into main match df
# Use left merge to keep all matches, filling 0 for 0-0 draws if any are missing from goals df
matches_full = pd.merge(matches_weather, goals_per_match, on='match_id', how='left')
matches_full['total_goals'] = matches_full['total_goals'].fillna(0)

# 4. Extract year from tournament ID for cleaner sorting
print("Data preprocessing complete.")
matches_full.head(3)
```

```python
# ------------------------------------------------------
# 3. Goal distribution
# ------------------------------------------------------
# Histogram of goals per match
plt.figure(figsize=(10, 6))
sns.countplot(
    data=matches_full, 
    x='total_goals', 
    hue='total_goals',     
    palette='Blues_d', 
    legend=False           
)

plt.title('Distribution of Goals per Match (All Tournaments)', fontsize=16)
plt.xlabel('Number of Goals', fontsize=12)
plt.ylabel('Count of Matches', fontsize=12)

# Calculate mean goals
mean_goals = matches_full['total_goals'].mean()
plt.axvline(mean_goals, color='red', linestyle='--', label=f'Mean Goals: {mean_goals:.2f}')
plt.legend()
plt.show()

print(f"Skewness of goals: {matches_full['total_goals'].skew()}")
```

```python
# Right-skewed distribution (skewness ~0.78) with an average of 2.51 goals 
# per match, most games feature 2-3 goals while high-scoring matches are 
# rare outliers. It resembles a Poisson distribution.
```

```python
# ------------------------------------------------------
# 4. Goals vs. Tournament
# ------------------------------------------------------
plt.figure(figsize=(12, 6))
# Boxplot to see the spread (median, quartiles) per tournament
sns.boxplot(data=matches_full, x='tournament_id', hue='tournament_id', y='total_goals', palette="Set2", legend=False)

plt.title('Goals per Match Distribution by Tournament', fontsize=16)
plt.xlabel('Tournament', fontsize=12)
plt.ylabel('Goals Scored', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Goals seems largely consistent across the tournaments, notable exception would
# be 2014, with higher median of 3, and greater variability
```

```python
# ------------------------------------------------------
# 5. Timing of goals, substitutions, and yellow cards
# ------------------------------------------------------
plt.figure(figsize=(14, 7))

# Plot 1: Goal timings
sns.kdeplot(goals['minute_regulation'], label='Goals', fill=True, color='green', alpha=0.3)

# Plot 2: Substitution timings
sns.kdeplot(substitutions['minute_regulation'], label='Substitutions', fill=True, color='#673AB7', alpha=0.3)

# Plot 3: Yellow card timings
# Filter only yellow cards
yellow_cards = bookings[bookings['is_yellow'] == True]
sns.kdeplot(yellow_cards['card_minute'], label='Yellow Cards', fill=True, color='yellow', alpha=0.3)

plt.title('Timeline Density: Goals vs. Subs vs. Cards', fontsize=16)
plt.xlabel('Match Minute', fontsize=12)
plt.ylabel('Density (Frequency)', fontsize=12)
plt.xlim(0, 120) # Limit to typical match time + extra time
plt.legend()
plt.show()
```

```python
# ------------------------------------------------------
# 6. Temperature distribution
# ------------------------------------------------------
plt.figure(figsize=(14, 6))

sns.violinplot(data=matches_full, x='tournament_id', hue='tournament_id', y='avg_temp', inner="quart", palette="coolwarm", legend=False)

plt.title('Temperature Distributions by Tournament', fontsize=16)
plt.ylabel('Average Match Temperature (Â°C)', fontsize=12)
plt.xlabel('Tournament', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------------------------------------------
# 7. Squad median age across tournaments
# ------------------------------------------------------
# 1. Convert rank_date to datetime objects
squads['rank_date'] = pd.to_datetime(squads['rank_date'], format='%Y-%m-%d')

# 2. Extract year
squads['tournament_year'] = squads['rank_date'].dt.year

# 3. Plot squad ages by tournament year
plt.figure(figsize=(12, 6))

sns.boxplot(
    data=squads, 
    x='tournament_year',
    hue='tournament_year',
    y='median_age', 
    palette="Purples",
    legend=False
)

plt.title('Distribution of Squad Median Ages by Tournament', fontsize=16)
plt.xlabel('Tournament Year', fontsize=12)
plt.ylabel('Median Squad Age', fontsize=12)
plt.show()
```

```python
# ------------------------------------------------------
# 8. Comparing red cards in group stage vs. knockout
# ------------------------------------------------------
logical_order = [
    "group stage", 
    "round of 16", 
    "quarter-finals", 
    "semi-finals", 
    "third-place match", 
    "final"
]
# 1. Total matches per stage per tournament
stage_match_counts = matches_weather.groupby(['tournament_id', 'stage_name'])['match_id'].nunique().reset_index(name='total_matches')

# 2. Red cards per stage per tournament
stage_red_cards = bookings[bookings['is_red'] == True].groupby(['tournament_id', 'stage_name']).size().reset_index(name='total_red_cards')

# 3. Merge dfs
# Left merge on match counts to ensure we keep stages even if they had 0 red cards
stage_stats = pd.merge(stage_match_counts, stage_red_cards, on=['tournament_id', 'stage_name'], how='left')

# Fill NaN values with 0 (for stages that had 0 red cards)
stage_stats['total_red_cards'] = stage_stats['total_red_cards'].fillna(0)

# 4. Calculate the metric
stage_stats['red_cards_per_match'] = stage_stats['total_red_cards'] / stage_stats['total_matches']

# 5. Plotting
plt.figure(figsize=(12, 6))
sns.barplot(
    data=stage_stats, 
    x='stage_name', 
    y='red_cards_per_match', 
    hue='tournament_id', 
    palette='Reds',
    order=logical_order
)

plt.title('Normalised Red Card Rate (Cards per Match)', fontsize=16)
plt.xlabel('Stage', fontsize=12)
plt.ylabel('Red Cards per Match', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Tournament', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# ------------------------------------------------------
# 9. Correlation heatmap of match stats
# ------------------------------------------------------
# 1. Prepare substitution counts
subs_per_match = substitutions.groupby('match_id').size().reset_index(name='total_subs')

# 2. Create "Master Stats" df
# Start with MatchBox data
match_stats = matchbox.copy()

# 3. Merge with Weather & Goals using Team Names + Tournament
# Use this merge to fetch the match_id, avg_temp, and total_goals
match_stats = pd.merge(
    match_stats, 
    matches_full[['tournament_id', 'home_team_name', 'away_team_name', 'match_id', 'avg_temp', 'total_goals']], 
    left_on=['tournament_id', 'hname', 'aname'],       # Keys in MatchBox
    right_on=['tournament_id', 'home_team_name', 'away_team_name'], # Keys in Matches Full
    how='left'
)

# 4. Merge the Subs
match_stats = pd.merge(match_stats, subs_per_match, on='match_id', how='left')
match_stats['total_subs'] = match_stats['total_subs'].fillna(0)

# 5. Calculate match totals
match_stats['total_shots'] = match_stats['hshots'] + match_stats['ashots']
match_stats['total_fouls'] = match_stats['hfouls'] + match_stats['afouls']
match_stats['total_yellows'] = match_stats['hyellowCards'] + match_stats['ayellowCards']
match_stats['total_reds'] = match_stats['hredCards'] + match_stats['aredCards']

# 6. Select columns
cols_of_interest = [
    'total_goals', 
    'total_shots', 
    'total_fouls', 
    'total_yellows', 
    'total_reds', 
    'total_subs', 
    'avg_temp',
    'hPossession'
]

readable_names = {
    'total_goals': 'Total Goals',
    'total_shots': 'Total Shots',
    'total_fouls': 'Fouls',
    'total_yellows': 'Yellow Cards',
    'total_reds': 'Red Cards',
    'total_subs': 'Substitutions',
    'avg_temp': 'Temperature',
    'hPossession': 'Home Possession'
}

# Calculate correlation on the renamed df
corr_matrix = match_stats[cols_of_interest].rename(columns=readable_names).corr()

# 7. Plotting
plt.figure(figsize=(10, 8))

# Mask upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix, 
    mask=mask, 
    annot=True, 
    fmt=".2f", 
    cmap='vlag', 
    center=0,
    vmin=-0.5, 
    vmax=0.5,
    square=True, 
    linewidths=1, 
    linecolor='white',
    cbar_kws={"shrink": .7},
    annot_kws={"size": 11} 
)

plt.title('Correlation of Match Intensity & Conditions', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right') # Rotate x-labels for readability
plt.yticks(rotation=0)
plt.show()
```

```python
# ------------------------------------------------------
# 10. Do extreme temperatures reduce the no. of goals?
# ------------------------------------------------------
plt.figure(figsize=(10, 6))

# Scatter plot with a regression line
sns.regplot(data=matches_full, x='avg_temp', y='total_goals', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title('Relationship between Temperature and Total Goals', fontsize=16)
plt.xlabel('Average Match Temperature (Â°C)', fontsize=12)
plt.ylabel('Total Goals in Match', fontsize=12)
plt.show()
```
